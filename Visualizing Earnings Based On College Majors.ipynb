# Guided Project: Visualizing Earnings Based On College Majors

In the guided project, we'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data.
Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more.
Using visualizations, we can start to explore questions from the dataset like:
- Do students in more popular majors make more money?
- How many majors are predominantly male? Predominantly female?
- Which category of majors have the most students?


**PRELIMINARY WORK**

Let's import the libraries we need and remove rows containing null values

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
print('- First row of the dataset:\n\n',recent_grads.iloc[0])
print('\n\n- First five rows of th dataset:\n\n',recent_grads.head())
print('\n\n- Last five rows of the dataset:\n\n',recent_grads.tail())
print('\n\n-Summary statistics for all of the numeric columns:\n\n', recent_grads.describe())
```

```python
# Number of rows pre-cleaning
raw_data_count = len(recent_grads.index)
# Let's drop rows with missing values
recent_grads = recent_grads.dropna()
# Number of row after cleaning
cleaned_data_count = len(recent_grads.index)

# Let's compare the 2 counts and get the number of rows with missing values
print('There were', raw_data_count - cleaned_data_count ,'row(s) with missing values.')
```

**PANDAS, SCATTER PLOTS**

Let's generate a scatter plots to explore the following relations between columns:
- `Sample_size` and `Median`
- `Sample_size ` and `Unemployment_rate`
- `Full_time` and `Median`
- `ShareWomen` and `Unemployment_rate`
- `Men` and `Median`
- `Women` and `Median`

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full_time ')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. ShareWomen ')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs. Men')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs. Women')
```

From the scatter plots above, we can say that:
- Students in more popular majors make more money
- Students that majored in the subjects that were majority females make less money 
- There is a link between the number of full-time employees and median salary as the salaries made by full-time employees are used to calculate the median salary

**PANDAS, BAR PLOTS**

Let's use bar plots to:
- to compare the percentages of women (`ShareWomen`) from the first ten rows and last ten rows of the recent_grads dataframe
- to compare the unemployment rate (`Unemployment_rate`) from the first ten rows and last ten rows of the recent_grads dataframe.

```python
# First 10 rows
ax1 = recent_grads.head(10).plot.bar(x='Major', y='ShareWomen', legend=False)
ax1.set_title('First 10 rows of ShareWomen column')
#Last 10 rows
ax2 = recent_grads.tail(10).plot.bar(x='Major', y='ShareWomen', legend=False)
ax2.set_title('Last 10 rows of ShareWomen column')
```

```python
# First 10 rows
ax1 = recent_grads.head(10).plot.bar(x='Major', y='Unemployment_rate', legend=False)
ax1.set_title('First 10 rows of Unemployment_rate column')
#Last 10 rows
ax2 = recent_grads.tail(10).plot.bar(x='Major', y='Unemployment_rate', legend=False)
ax2.set_title('Last 10 rows of Unemployment_rate column')
```
